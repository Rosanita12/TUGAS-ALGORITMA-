```python
import numpy as np
import matplotlib.pyplot as plt

# Menentukan Titik dan Menghitung Gradien Garis Pertama
# Persamaan garis pertama melewati titik (e,f) dan (g,h)
e_10=1 #Titik (e,f)
f_10=1
g_10=2 #Titik (g,h)
h_10=2
m1_10=(h_10-f_10)/(g_10-e_10)
print ("Gradien garis pertamanya {}".format(m1_10))
print("Persamaan garis pertama adalah y={}x+{}".format(m1_10, -m1_10*e_10+f_10))
#y1 = m1*(x-e)+f

# Menghitung Titik dan Menghitung Gradien Garis Kedua
# Persamaan garis pertama kedua melewati titik (i,j) dan (k,l)
i_10=1 #Titik（i,j）
j_10=1
k_10=2 #Titik（k,l）
l_10=-2
m2_10=(l_10-j_10)/(k_10-i_10)
print( "Gradien garis ke duanya {}".format(m2_10))
print("Persamaan garis kedua adalah y={}x+{}".format(m2_10, -m2_10*i_10+j_10))
#y2=m2*(x-i)+j

# Menentukan Titik Potong Kedua Garis
if m1_10==m2_10:
  print( "OMG karena Gradiennya sama, jadi kemungkinan persamaannya tidak ada solusi atau solusinya ada banyak.")
else:
    # Titik potong kedua garis pada (tp_x, tp_y)
    tp_x_10 = (m1_10*e_10-f_10-m2_10*i_10+j_10)/(m1_10-m2_10)
    tp_y_10 = m2_10*(tp_x_10-i_10)+j_10

    print ("Titik potong untuk x dan y adalah ({},{})".format(tp_x_10,tp_y_10))

# Plotting Garis dan Titik Potong
    x_10=np.arange(tp_x_10-5,tp_x_10+5,0.01) #Membuat array x untuk plotting
    plt.figure(figsize=(4,4),dpi=150) #Menentukan ukuran dan resolusi figure

    plt.plot(x_10,m1_10*(x_10-e_10)+f_10,color='red',label='y={}x+{}'.format(m1_10,-m1_10*e_10+f_10)) #Plot garis pertama
    plt.plot(x_10,m2_10*(x_10-i_10)+j_10,color='blue',label='y={}x+{}'.format(m2_10,-m2_10*i_10+j_10)) #Plot garis kedua
    plt.plot(tp_x_10,tp_y_10,color='green',marker='o',markersize=4) #Plot titik potong
    plt.legend(loc="best") #Menampilkan legend
# Menampilkan hasil
    plt.show()
print('Thank You')
```
