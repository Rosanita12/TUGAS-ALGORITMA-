```python
print("Project 1")
#Persamaan Garis dengan ax^2+bx+c=0
# Masukkan nilai a,b,c
# Suatu persamaan yang mencari akar akar dengan rumus abc dan faktorilasi
# Dalam rumus a,b,c terdapat diskriminan yang dapat dicari dengan rumus D=b^2-4ac
# Terdapat 3 kemungkinan akar akar yang terbentuk berdasarkan nilai dari diskriminan
#1.Jika D<0;akar imajiner
#2.Jika D=0;akar kembar
#3.Jika D>0;akar real berbeda
# Berdasarkan nilai a dari persamaan kuadrat ax^2+bx+c=0 dapat terlihat kurva dari grafiknya
#1.Jika a > 0;kurva membuka keatas
#2.Jika a < 0;kurva membuka kebawah
# Apabila yang diketahui akar akar dari suatu persamaan kuadrat , maka dapat dicari persamaan kuadratnya dengan (x^2-(a+b)x+ab=0)
#Langkah-langkah menentukan akar-akar persamaan kuadrat
# 1. Memanggil numpy
import numpy as np
# 2. Memanggil pyplot
import matplotlib.pyplot as plt
# 3. input untuk nilai a,b,c dalam persamaan ax^+bx+c=0
o_10 = float(input("inputlah nilai dari a "))
c_10 = float(input("inputlah nilai dari b "))
a_10 = float(input("inputlah nilai dari c "))
if o_10 != 0:
  print("Persamaan kuadrat yang terbentuk ")
else:
  print("Tidak bisa terbentuk persamaan kuadrat")
print("Adapun persamaan kuadrat yang terbentuk adalah ({}x**2+{}x+{})".format(o_10,c_10,a_10))
print("Adapun persamaan Kuadratnya adalah ({}x**2+{}x+{})".format(o_10,c_10,a_10))
# 4. Input rumus diskriminan
D_10 = (c_10**2-4*o_10*a_10)
# 5. Input rumus x1 akar 1,x2 akar 2
import cmath
x1_10 = (-c_10+cmath.sqrt(D_10))/(2*o_10)
x2_10 = (-c_10-cmath.sqrt(D_10))/(2*o_10)
# 6. Menunjukkan hasil
print("Adapun akar-akar dari persamaannya adalah ({:.1f},{:.1f})".format(x1_10,x2_10))

print("================================== Terima Kasih ==================================")
```

```python
print("Project 2")
# 1.Memanggil numpy
import numpy as np
# 2.Memanggil pyplot
import matplotlib.pyplot as plt
# 3.Memanggil cmath
import cmath
# 4.Masukkan untuk nilai a,b,c
a_10 = float(input("input  nilai dari a "))
b_10 = float(input("input nilai dari b "))
c_10 = float(input("input nilai dari c "))
if a_10 != 0:
  print(" MakaPersamaan kuadrat yang terbentuk adalah")
else:
  print("Jadi Tidak dapat membentuk persamaan kuadrat")
print("Pjadi ersamaan kuadrat yang dibentuk ({}x**2+{}x+{})".format(a_10,b_10,c_10))
print("Jadi Persamaan Kuadratnya adalah ({}x**2+{}x+{})".format(a_10,b_10,b_10))
# 5.Masukkan rumus diskriminan
D_10 = (b_10**2-4*a_10*b_10)
if D_10 == 0:
  print("Akar-akarnya kembar")
# 6.Masukkan rumus x1,x2
  x1_10 = (-b_10+cmath.sqrt(D_10))/(2*a_10)
  x2_10 = (-b_10-cmath.sqrt(D_10))/(2*a_10)
  print("Akar-akar dari persamaannya adalah ({:.1f},{:.1f})".format(x1_10,x2_10))
elif D_10 > 0:
  print("Akar real berbeda")
  x1_10 = (-b_10 + cmath.sqrt(D_10))/(2*a_10)
  x2_10 = (-b_10 - cmath.sqrt(D_10))/(2*a_10)
  print("Akar-akarnya adalah ({:.1f},{:.1f})".format(x1_10,x2_10))
elif D_10 < 0:
  print("Akarnya imajiner")
  x1_10 = (-b_10 + cmath.sqrt(D_10))/(2*a_10)
  x2_10 = (-b_10 - cmath.sqrt(D_10))/(2*a_10)
  print("Akar-akarnya adalah ({:.1f},{:.1f})".format(x1_10,x2_10))
else:
  print("TJadi tidak terdapat solusi pada penyelesaian ini ")

print("================================== Terima Kasih ==================================")
```

```python
print("============================== Project 3 ==============================  ")
# Grafik pada persamaan kuadrat
# Langkah-langkahnya adalah :
# 1. Memanggil numpy
import numpy as np
# 2. Memanggil pyplot
import matplotlib.pyplot as plt
# 3. Memanggil cmath
import cmath
# 4. Masukkan nilai a, b dan c dalam persamaan ax**2 + bx + c = 0
a_10 = float(input("Nilai konstanta a adalah: "))
b_10 = float(input("Nilai konstanta b adalah: "))
c_10 = float(input("Nilai konstanta c adalah: "))

if a_10 == 0:
    print("Error, nilai yang anda input bukan persamaan kuadrat")
else:
    print("Nilai yang anda input bisa membentuk persamaan kuadrat")
    print("Persamaan Kuadratnya adalah ({}x**2 + {}x + {})".format(a_10, b_10, c_10))

    # 5. Masukkan rumus diskriminan
    D_10 = b_10**2 - 4*a_10*c_10
    if D_10 == 0:
        print("Wahh, Akar-akarnya kembar")
    elif D_10 > 0:
        print("Akar real berbeda")
    else:
        print("Akarnya imajiner")

    # 6. Masukkan rumus x1, x2
    x1_10 = (-b_10 + cmath.sqrt(D_10)) / (2*a_10)
    x2_10 = (-b_10 - cmath.sqrt(D_10)) / (2*a_10)
    print("Akar-akarnya adalah ({:.2f}, {:.2f})".format(x1_10.real, x2_10.real))

    # 7. Format garis
    x_10 = np.linspace(-10, 10, 400)
    y_10 = a_10*x_10**2 + b_10*x_10 + c_10
    plt.figure(figsize=(8, 6))
    plt.plot(x_10, y_10, label=f'{a_10}x^2 + {b_10}x + {c_10} = 0')

    if D_10 >= 0:
        plt.scatter([x1_10.real, x2_10.real], [0, 0], color='red')
        plt.text(x1_10.real, 0.5, f'({x1_10.real:.2f}, 0)', fontsize=12, ha='center')
        plt.text(x2_10.real, 0.5, f'({x2_10.real:.2f}, 0)', fontsize=12, ha='center')
    else:
        plt.scatter([x1_10.real, x2_10.real], [x1_10.imag, x2_10.imag], color='red')
        plt.text(x1_10.real, x1_10.imag, f'({x1_10.real:.2f}, {x1_10.imag:.2f})', fontsize=12, ha='center')
        plt.text(x2_10.real, x2_10.imag, f'({x2_10.real:.2f}, {x2_10.imag:.2f})', fontsize=12, ha='center')

    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(color='gray', linestyle='--', linewidth=0.5)
    plt.title('Grafik Persamaan Kuadrat')
    plt.xlabel('x_10')
    plt.ylabel('y_10')
    plt.legend()
    plt.show()

    print("Terima Kasih ")
```

```python
print("Menentukan Titik Puncak dan Sumbu Simetri")
# 1. Memanggil numpy
import numpy as np
# 2. Memanggil pyplot
import matplotlib.pyplot as plt
# 3. Memanggil cmath
import cmath
# 4. Masukkan nilai a,b,c
a_10 = float(input("Masukan Nilai a : "))
b_10 = float(input("Masukan Nilai b : "))
c_10 = float(input("Masukan Nilai c : "))

def hitung_titik_puncak_dan_sumbu_simetri(a_10, b_10, c_10):
    # Sumbu simetri (x) adalah -b / (2a)
    sumbu_simetri = -b_10 / (2 * a_10)

    # Titik puncak (x, y)
    # x dari titik puncak adalah sumbu simetri
    # y dari titik puncak adalah nilai fungsi kuadrat di x = sumbu_simetri
    titik_puncak_x_10 = sumbu_simetri
    titik_puncak_y_10 = a_10 * (titik_puncak_x_10 ** 2) + b_10 * titik_puncak_x_10 + c_10

    return (titik_puncak_x_10, titik_puncak_y_10), sumbu_simetri

def plot_persamaan_kuadrat(a_10, b_10, c_10):
    # Menghitung titik puncak dan sumbu simetri
    titik_puncak, sumbu_simetri = hitung_titik_puncak_dan_sumbu_simetri(a_10, b_10, c_10)

    # Membuat rentang x untuk plot
    x_10 = np.linspace(titik_puncak[0] - 10, titik_puncak[0] + 10, 400)
    y_10 = a_10 * x_10**2 + b_10 * x_10 + c_10

    # Plot grafik persamaan kuadrat
    plt.figure(figsize=(10, 6))
    plt.plot(x_10, y_10, label=f'{a_10}x^2 + {b_10}x + {c_10}')

    # Plot titik puncak
    plt.plot(titik_puncak[0], titik_puncak[1], 'ro')  # Titik puncak berwarna merah
    plt.annotate(f'Titik puncak\n({titik_puncak[0]:.2f}, {titik_puncak[1]:.2f})',
                 xy=titik_puncak,
                 xytext=(titik_puncak[0] + 1, titik_puncak[1] + 10),
                 arrowprops=dict(facecolor='black', shrink=0.05))

    # Plot sumbu simetri
    plt.axvline(x=sumbu_simetri, color='gray', linestyle='--', label='Sumbu simetri')

    # Menambahkan label dan judul
    plt.xlabel('x_10')
    plt.ylabel('y_10')
    plt.title('Titik Puncak Grafik Persamaan Kuadrat')
    plt.axvline(x=0,color='black')
    plt.axhline(y=0,color='black')
    plt.legend()
    plt.grid(True)
    plt.show()

# Menjalankan fungsi plot_persamaan_kuadrat dengan nilai a, b, dan c yang diberikan
plot_persamaan_kuadrat(a_10, b_10, c_10)
print("================================== Terima Kasih ==================================")
```

```python
print("Menentukan Posisi Titik Puncak")
# Menentukan posisi titik puncak (di kiri sumbu y *(pada sumbu y b=0), dikanan sumbu y)
# 1. Memanggil numpy
import numpy as np
# 2. Memanggil pyplot
import matplotlib.pyplot as plt
# 3. Masukkan nilai a, b, c
a_10 = float(input("Input Nilai a : "))
b_10 = float(input("Input Nilai b : "))
c_10 = float(input("Input Nilai c : "))

def hitung_titik_puncak_dan_sumbu_simetri(a_10, b_10, c_10):
    # Sumbu simetri (x) adalah -b / (2a)
    sumbu_simetri = -b_10 / (2 * a_10)

    # Titik puncak (x, y)
    # x dari titik puncak adalah sumbu simetri
    # y dari titik puncak adalah nilai fungsi kuadrat di x = sumbu_simetri
    titik_puncak_x_10 = sumbu_simetri
    titik_puncak_y_10 = a_10 * (titik_puncak_x_10 ** 2) + b_10 * titik_puncak_x_10 + c_10

    return (titik_puncak_x_10, titik_puncak_y_10), sumbu_simetri

def tentukan_posisi_titik_puncak(titik_puncak_x_10):
    if titik_puncak_x_10 < 0:
        posisi = "di kiri sumbu y"
    elif titik_puncak_x_10 > 0:
        posisi = "di kanan sumbu y"
    else:
        posisi = "pada sumbu y"
    return posisi

def plot_persamaan_kuadrat(a_10, b_10, c_10):
    # Menghitung titik puncak dan sumbu simetri
    titik_puncak, sumbu_simetri = hitung_titik_puncak_dan_sumbu_simetri(a_10, b_10, c_10)

    # Menentukan posisi titik puncak
    posisi = tentukan_posisi_titik_puncak(titik_puncak[0])

    # Membuat rentang x untuk plot
    x_10 = np.linspace(titik_puncak[0] - 10, titik_puncak[0] + 10, 400)
    y_10 = a_10 * x_10**2 + b_10 * x_10 + c_10

    # Plot grafik persamaan kuadrat
    plt.figure(figsize=(10, 6))
    plt.plot(x_10, y_10, label=f'{a_10}x^2 + {b_10}x + {c_10}')

    # Plot titik puncak
    plt.plot(titik_puncak[0], titik_puncak[1], 'ro')  # Titik puncak berwarna merah
    plt.annotate(f'Titik puncak\n({titik_puncak[0]:.2f}, {titik_puncak[1]:.2f})\n{posisi}',
                 xy=titik_puncak,
                 xytext=(titik_puncak[0] + 1, titik_puncak[1] + 10),
                 arrowprops=dict(facecolor='purple', shrink=0.05))

    # Plot sumbu simetri
    plt.axvline(x=sumbu_simetri, color='yellow', linestyle='--', label='Sumbu simetri')

    # Menambahkan label dan judul
    plt.xlabel('x_10')
    plt.ylabel('y_10')
    plt.title('Posisi Titik Puncak Grafik Persamaan Kuadrat')
    plt.axvline(x=0,color='purple')
    plt.axhline(y=0,color='purple')
    plt.legend()
    plt.grid(True)
    plt.show()

# Menjalankan fungsi plot_persamaan_kuadrat dengan nilai a, b, dan c yang diberikan
plot_persamaan_kuadrat(a_10, b_10, c_10)
print("================================== Terima Kasih ==================================")
```
