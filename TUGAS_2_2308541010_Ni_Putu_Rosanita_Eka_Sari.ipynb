```python
print ("SELAMAT DATANG DI PROGRAM MENGHITUNG VOLUME BALOK")
p_10 = float (input ("Berapa panjang balok?:"))
l_10 = float (input ("Berapa lebar balok?:"))
t_10 = float (input ("Berapa tinggi balok?:"))

volume_10 = float(p_10)*float(l_10)*float(t_10)

print ("=========================================")
print (f" Jadi luas balok tersebut adalah {volume_10} cm^3")
```

```python
print ("BINGUNG MAU MENGHITUNG VOLUME TABUNG DAN LUAS PERMUKAAN TABUNG")

r_10 = float (input("Masukan jari-jari tabung:"))
t_10 = float (input ("Masukan tinggi tabung: "))

Lp_10 = (2 * 22/7 * r_10) *(r_10+ t_10)
V_10 = (22/7* r_10* r_10 * t_10)
                            
print ("========================================================================================================")
print (f"Jadi Volume Tabung adalah {V_10} cm^3 dan Luas Permukaan Tabung adalah {Lp_10} cm^3" )
```

```python
print ("Ini adalah kalkulator ajib, kamu tinggal memasukan angka dan hasil dari +,-,*,/, dan ** akan muncul ")
a1_10 = input ("Masukan angka 1: ")
a2_10 = input ("Masukan angka 2: ")

sub_10 = float(a1_10) - float(a2_10)
inc_10 = float(a1_10)+ float(a2_10)
multi_10 = float(a1_10) * float(a2_10)
bagi_10 = float(a1_10)/float(a2_10)
pangkat_10 = float(a1_10)** float(a2_10)
print ("===============================================")
print (f"Hasil pengurangan adalah {sub_10}")
print (f"Hasil pertambahan adalah {inc_10}")
print (f"Hasil perkalian adalah {multi_10}")
print (f"Hasil pembagian adalah {bagi_10}")
print (f"Hasil pangkat adalah {pangkat_10}")
```
